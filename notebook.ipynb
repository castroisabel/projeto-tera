## Parte I: Estrutura do Projeto
 
### Contexto:
 
No Brasil, o total de operações feitas em todos os canais bancários chegou a 103,5 bilhões em 2020, segundo dados da Pesquisa Febraban de Tecnologia Bancária 2021. Como consequência deste cenário, existe um grande número de golpes aplicados através dessas transações - principalmente no que tange golpes relacionados às transações via cartão de crédito. Segundo dados da ABECS (Associação Brasileira das Empresas de Cartões de Crédito), os brasileiros transacionaram 1,6 trilhão de reais por meio de cartões de crédito em 2021, onde neste valor se concentra grande parte das transações do varejo brasileiro. 
 
Frequentemente, as fraudes são resultado de descuidos ou incidentes por parte dos usuários, entretanto as ferramentas de segurança e os analistas de prevenção a fraude das instituições financeiras são responsáveis por garantir que compras indevidas ou suspeitas não sejam aprovadas. Dessa forma, os bancos vêm tomando diversas e constantes medidas com o intuito de fortalecer suas infraestruturas internas, seus aplicativos e demais canais digitais, além de apostar em novas tecnologias e campanhas de conscientização do cliente.
 
Os desafios são gigantescos e a velocidade em que o cenário se altera é muito maior do que a de criação de estratégias eficientes para resguardar os clientes que realizam compras em ambientes digitais. Em resposta a isso, o investimento em modelos que identifiquem de forma precisa uma transação inválida é fundamental não só para evitar prejuízos financeiros, mas também para evitar o constrangimento de usuários que possuem seu cartão bloqueado no momento de uma compra por suspeita de fraude.
 
Desse modo, é importante que as empresas sejam capazes de reconhecer transações fraudulentas para melhorar a experiência do cliente, além de reduzir a perda de receita financeira e aumentar a sua credibilidade frente ao mercado. 
 
### Problema do negócio:
 
Existem diferentes abordagens para solucionar o problema de compras indevidas em cartões de crédito. Utilizando algoritmos não supervisionados, é possível analisar a última compra feita e compará-la com o histórico de compras de um determinado usuário. Valores, localização, moeda, horário, estabelecimento de compra, forma de pagamento e muitos outros fatores. Nesse método, é possível avaliar e agrupar clientes com perfis de compras semelhantes, utilizando esses padrões para tomar uma decisão a partir de comportamentos que são considerados suspeitos. Em contrapartida, um modelo supervisionado que busca apenas uma resposta como se é de fato fraude ou não, pode ser mais simples e eficiente computacionalmente e em termos de infraestrutura.
 
Diante das pesquisas e dos números apresentados, nosso projeto busca justamente abordar o problema de fraudes em transações usando métodos supervisionados para construir modelos que classifiquem uma transação financeira realizada por meio de um cartão de crédito como “fraude” ou “não fraude”.
 
### Impacto:
 
Tendo em vista o período atual e as partes envolvidas nas transações financeiras, os varejistas online são os mais afetados pelos fraudadores, por permitirem transações de cartão não presente (CNP). Os prejuízos vão de perda do investimento na aquisição daquele cliente e passam pela perda de confiança dos consumidores, podendo levar até a rescisão de contratos por altos índices de Chargeback.
 
Nosso modelo de detecção visa identificar com precisão as transações fraudulentas com o mínimo de transações legítimas barradas (falsos positivos), enquanto levamos em conta os custos de uma possível fraude em comparação com o custo da perda de um cliente, evitando assim prejuízos para os estabelecimentos, protegendo a reputação das empresas participantes na transação e a perda de credibilidade, ao não gerar clientes detratores.
 
### Desenho da solução:
 
Neste trabalho, temos o objetivo de criar um modelo de previsão que consiga identificar casos de fraude em transações de cartão de crédito. Isso significa que, a partir das variáveis que caracterizam uma transação, retornaremos uma classificação onde julgamos um evento como fraude ou não.  Para isso, utilizaremos um conjunto de dados fictício de transações de cartão de crédito onde estão contidas informações pessoais de clientes, especificações sobre a transação, etc.
 
A princípio, o projeto será modelado como um problema de classificação binária, ou seja, quando precisamos identificar a que classe uma determinada amostra pertence. Nesse caso, possuímos apenas duas classes: transação válida e transação inválida. Desse modo, o objetivo desse projeto é construir modelos de aprendizado de máquina implementando algoritmos supervisionados, nesse caso iremos selecionar uma amostra de treinamento que contenha o rótulo correto correspondente a cada transação e tentaremos reconhecer padrões de comportamento que são considerados suspeitos e indicativos de fraude para a tomada de decisão.
 
Após o treinamento do modelo, serão realizados testes em novas amostras, visando validar os modelos construídos. Nessa etapa, procuraremos aplicar o modelo em dados reais utilizando técnicas de análise de componentes principais. Por fim, criaremos uma aplicação que nos permita consumir do nosso modelo e receber requisições em um expectativa de simular o modelo em produção.

```python
def pre_processing(data):
  data = data.copy()
  data.dropna(inplace=True)
  data.drop_duplicates(inplace=True)
  data['trans_date_trans_time'] = pd.to_datetime(data['trans_date_trans_time'])
  data['dob'] = pd.to_datetime(data['dob'])
  data[['cc_num']] = data[['cc_num']].astype('object')
  data['name'] = data['first'] + ' ' + data['last']
  data['age'] = np.round((data['trans_date_trans_time'] - data['dob'])/np.timedelta64(1, 'Y'))
  data['hour'] = pd.to_datetime(data['trans_date_trans_time']).dt.hour
  data['day'] = pd.to_datetime(data['trans_date_trans_time']).dt.dayofweek
  data['month'] = pd.to_datetime(data['trans_date_trans_time']).dt.month
  data.drop(['first', 'last', 'trans_num', 'dob'], axis=1, inplace = True)
  data.sort_values(by='trans_date_trans_time', inplace=True)

  return data
  
df = pre_processing(raw_data)
```

Processamento inicial dos dados:
- Removemos amostras duplicadas
- Modificamos o tipo de algumas variáveis
- Juntamos a coluna `fisrt` e a coluna `last` em uma só coluna chamada `name`
- Criamos a nova coluna `age`
- Excluimos colunas consideradas irrelevantes
- Ordenamos os dados em ordem crescente de data da transação

- **Hipótese:** O valor transacionado em transações fraudulentas é estatisticamente diferente (maior) do valor transacionado em transações autênticas.

**Teste de Hipótese**:

1. $p-value < 0.05$: $H_0$ rejeitada

2. $H_0$: não existe diferença entre o valor transacionado em transações normais e fraudulentas

3. $H_A$: existe diferença entre o valor transacionado em transações normais e fraudulentas

4. Teste de Kolmogorov-Smirnov (KS)
  - É utilizado na comparação de dois grupos independentes;
  - Verifica a concordância entre duas distribuições acumuladas.

```python
# Teste de hipótese
from scipy.stats import ks_2samp

print('Reject the null hypothesis: {}'.format(ks_2samp(df.query('is_fraud==0').amt, df.query('is_fraud==1').amt)))
```

```python
# Treinamento de uma Floresta Aleatória
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import confusion_matrix, classification_report

model = RandomForestClassifier(random_state=42, n_estimators=50)
model.fit(X_resampled, y_resampled)
```

```python
# Análise do threshold
predicted_proba = model.predict_proba(X_resampled)

threshold_search = {}
threshold = np.linspace(0.3, 0.8, num=51).round(2)

for t in threshold:
  predicted = (predicted_proba [:,1] >= t).astype('int')
  cm = confusion_matrix(y_true=y_resampled, y_pred=predicted)
  sum = cm[0][1] + cm[1][0]
  threshold_search[t] = []
  threshold_search[t].append(sum)

def argmin(d):
    if not d: return None
    min_val = min(d.values())
    return [k for k in d if d[k] == min_val][0]

min_index = argmin(threshold_search)
print('The best threshold: ', min_index)
```

```python
# Métricas
threshold = min_index

predicted_proba = model.predict_proba(X_test)
predicted = (predicted_proba [:,1] >= threshold).astype('int')
print('Classification report:\n', classification_report(y_test, predicted))

cm = confusion_matrix(y_true=y_test, y_pred=predicted)
print('Confusion matrix:\n', cm)
print('\nShare of Non-Fraud in Test Data:', 1-round(y_test.sum()/len(y_test),4))
```
